## **Correlaciones**

La correlación mide la relación entre dos variables, es decir, cómo cambian juntas. Si dos variables tienen una correlación positiva, significa que cuando una aumenta, la otra también lo hace. Si tienen una correlación negativa, cuando una aumenta, la otra disminuye.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
url = 'https://raw.githubusercontent.com/conectivaoficial/analisis-tech/refs/heads/main/estudio.csv'
data=pd.read_csv(url, sep=",")
data.head()
```

```python
# Calculamos la matriz de correlaciones
correlaciones = data.corr()

# Mostramos las correlaciones
print("Matriz de correlación:")
correlaciones
```

```python
# Visualizamos la matriz de correlación usando un mapa de calor
plt.figure(figsize=(8, 6))
plt.title("Matriz de Correlación")
plt.imshow(correlaciones, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(np.arange(5), data.columns, rotation=45)
plt.yticks(np.arange(5), data.columns)
plt.show()
```

```python
# Creamos el scatter plot entre las dos variables seleccionadas
plt.figure(figsize=(8, 6))
plt.scatter(data['Horas de Estudio'], data['Calificación Examen'], color='blue')

# Títulos y etiquetas
plt.title('Horas de Estudio vs Calificación Examen')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación Examen')

# Mostrar gráfico
plt.show()
```

```python
# Crea los scatter plots para las relaciones entre variables
fig, axes = plt.subplots(2, 3, figsize=(15, 10))  # 2 filas, 3 columnas de gráficos

# Scatter plot 1: Horas de Estudio vs Calificación Examen
axes[0, 0].scatter(data['Horas de Estudio'], data['Calificación Examen'], color='blue')
axes[0, 0].set_title('Horas de Estudio vs Calificación Examen')
axes[0, 0].set_xlabel('Horas de Estudio')
axes[0, 0].set_ylabel('Calificación Examen')

# Scatter plot 2: Horas de Estudio vs Edad
axes[0, 1].scatter(data['Horas de Estudio'], data['Edad'], color='green')
axes[0, 1].set_title('Horas de Estudio vs Edad')
axes[0, 1].set_xlabel('Horas de Estudio')
axes[0, 1].set_ylabel('Edad')

# Scatter plot 3: Horas de Estudio vs Horas de TV
axes[0, 2].scatter(data['Horas de Estudio'], data['Horas de TV'], color='red')
axes[0, 2].set_title('Horas de Estudio vs Horas de TV')
axes[0, 2].set_xlabel('Horas de Estudio')
axes[0, 2].set_ylabel('Horas de TV')

# Scatter plot 4: Calificación Examen vs Horas de TV
axes[1, 0].scatter(data['Calificación Examen'], data['Horas de TV'], color='purple')
axes[1, 0].set_title('Calificación Examen vs Horas de TV')
axes[1, 0].set_xlabel('Calificación Examen')
axes[1, 0].set_ylabel('Horas de TV')

# Scatter plot 5: Edad vs Tareas Entregadas
axes[1, 1].scatter(data['Edad'], data['Tareas Entregadas'], color='orange')
axes[1, 1].set_title('Edad vs Tareas Entregadas')
axes[1, 1].set_xlabel('Edad')
axes[1, 1].set_ylabel('Tareas Entregadas')

# Scatter plot 6: Horas de TV vs Tareas Entregadas
axes[1, 2].scatter(data['Horas de TV'], data['Tareas Entregadas'], color='brown')
axes[1, 2].set_title('Horas de TV vs Tareas Entregadas')
axes[1, 2].set_xlabel('Horas de TV')
axes[1, 2].set_ylabel('Tareas Entregadas')

# Ajuste de disposición de gráficos
plt.tight_layout()

# Mostrar los gráficos
plt.show()
```

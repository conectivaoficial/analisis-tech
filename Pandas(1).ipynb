### **Introducción a Pandas:**

Pandas es una poderosa librería de Python diseñada para el análisis de datos y la manipulación de datos estructurados. Fue creada por Wes McKinney en 2008 y se ha convertido en una herramienta esencial en el campo de la ciencia de datos y la ingeniería de datos.

## **¿Qué es Pandas?**

Pandas es una librería que proporciona estructuras de datos y herramientas de análisis de datos fáciles de usar en Python. Su nombre proviene de "Panel Data" (datos de panel), una referencia a su capacidad para manejar datos estructurados en forma de tablas o DataFrames, similar a una hoja de cálculo de Excel. Pandas es especialmente útil para tareas como limpieza, manipulación, análisis y visualización de datos.

### **Ventajas de usar Pandas en análisis de datos:**

*Las ventajas de utilizar Pandas en el análisis de datos incluyen:*

1. Manejo de Datos Estructurados: Pandas proporciona DataFrames, una estructura de datos tabular que permite trabajar con datos estructurados de manera eficiente.

2. Eficiencia en la Manipulación: Pandas es eficiente para operaciones de limpieza y transformación de datos, como filtrado, selección, agrupación y pivotado.

3. Integración con Datos Externos: Permite la importación y exportación de datos desde y hacia varios formatos de archivo, como CSV, Excel, SQL, y más.

4. Manipulación de Fechas y Tiempos: Ofrece herramientas para trabajar con fechas y horas de manera sencilla.

5. Indexación Personalizada: Permite etiquetar y organizar los datos mediante índices personalizados.

6. Manejo de Datos Faltantes: Pandas facilita la detección y el tratamiento de valores nulos o faltantes en los conjuntos de datos.

7. Funciones Estadísticas y de Agregación: Proporciona herramientas para calcular estadísticas descriptivas y realizar operaciones de agregación en los datos.

**Estructuras de Datos en Pandas:**

Las dos estructuras de datos principales en Pandas son:

1. **DataFrame**: Un DataFrame es una estructura de datos bidimensional que se asemeja a una tabla de base de datos o una hoja de cálculo de Excel. Consiste en filas y columnas, donde cada columna puede contener diferentes tipos de datos. Los DataFrames son especialmente útiles para el análisis de datos tabulares.

2. **Series**: Una Serie es una estructura de datos unidimensional similar a una matriz o una columna en una hoja de cálculo. Cada Serie tiene un índice que permite etiquetar y acceder a los elementos. Puedes pensar en una Serie como una columna de un DataFrame.

Ejemplo de creación de un DataFrame en Pandas:

```python
import pandas as pd
```

```python
datos=[["Helena", "F", "Bucaramanga", "13"],
       ["Esteban", "M", "Medellin", "24"],
       ["Edith", "F","Bucaramanga","59"],
       ["Jairo","M","Cúcuta","60"],
       ["José","M","Cali","45"]]
```

```python
datos
```

## **Series: una dimensión de datos etiquetados:**

Una Serie en Pandas es una estructura de datos unidimensional que puede contener datos de diferentes tipos, pero a diferencia de una lista de Python, cada elemento en una Serie está etiquetado con una etiqueta o índice. Estos índices son útiles para acceder y referirse a los datos de manera más intuitiva.

Ejemplo de creación de una Serie en Pandas:

```python
#Creacion de una serie a partir de una lista
mi_serie=pd.Series([10,20,30,40,50,60], index=["a","b","c","d","e","f"])
mi_serie
```

**DataFrames: tablas bidimensionales de datos:**

Un DataFrame es una estructura de datos bidimensional en Pandas que se asemeja a una tabla de base de datos o una hoja de cálculo de Excel. Consiste en filas y columnas, donde cada columna puede contener diferentes tipos de datos. Los DataFrames son ideales para trabajar con datos tabulares y facilitan el análisis y la manipulación de datos.

```python
#Index y titulos
titulos=["Nombre","Genero","Ciudad","Edad"]
indices=["uno","dos","tres","cuatro","cinco"]

#Creamos el dataframe df a partir de los datos y usamos los indices y titulos
df=pd.DataFrame(datos, index=indices, columns=titulos)
df
```

```python
df.columns  #Ver nombres de las columnas existentes en df
```

## **Manipulación de Datos con Pandas:**

Pandas ofrece una amplia gama de funciones para la manipulación de datos, que incluye:

**Agregar columnas:**

```python
df["Teléfono"]=["3103103103","3113113113","31231231233","3213213213","3153153155"]
df
```

```python
#Cambio de nombre en columnas
df.columns=["Name","Gender","City","Age","Phone"]
df
```

```python
df.columns=["Nombre","Genero","Ciudad","Edad","Teléfono"]
```

**Tipos de datos:**

```python
df["Edad"].dtype
```

```python
df["Edad"]=pd.to_numeric(df["Edad"])
df["Edad"].dtype
```

**Filtrado de datos:**

```python
df[(df["Edad"]<30)]   #Filtramos para ver la tabla con los datos de personas menores de 30


# df el primer df, representa el DataFrame completo (la tabla de datos).
# df["Edad"], extrae la columna "Edad" del DataFrame. Esto da como respuesta una serie es decir, un objeto de pandas que representa una sola columna.
```

```python
df[(df["Ciudad"]=="Bucaramanga")] #Filtramos para ver personas que esten en Bucaramanga
```

```python
df[(df["Edad"]<30)  & (df["Ciudad"]=="Bucaramanga")]  #Filtramos doble, personas que tienen menos de 30 años y estan en bBUcaramanga
```

```python
ciudad=input("Ingresa por favor la ciudad de donde deseas consultar: ").capitalize()  #Busquemos valores en una columna
df[df.Ciudad.str.contains(ciudad)]
```

**Modificación de datos:**

Puedes modificar los datos en un DataFrame de Pandas de varias maneras, como la asignación directa de valores a una columna o aplicando funciones a los datos existentes:

```python
df["Teléfono"]=df["Teléfono"].replace("31231231233","3123123123")
df
```

```python
df["Ciudad"]=df["Ciudad"].replace("Medellin","Medellín")
df
```

```python
# Aplicación de una función anónima a una columna
df["Edad"]=df["Edad"].apply(lambda x: x+1)
df


# Una función anónima es una función que no tiene un nombre explícito, y en Python se crean utilizando la palabra clave lambda.
# Por esta razón, también se las conoce como funciones lambda.

# .apply(lambda x: x + 1)
    # .apply(): Es un método de pandas que permite aplicar una función sobre cada valor individual en una columna (o serie).
    # lambda x: x + 1: Es una función anónima (lambda function) que toma un valor x como entrada y devuelve x + 1.
    # Básicamente, la función:
        # Recibe cada valor de la columna "Edad".
        # Suma 1 a cada valor.
        # Devuelve el resultado.
        # Para el ejemplo anterior:
           # x = 13 → 13 + 1 = 14
           # x = 24 → 24 + 1 = 25
           # x = 59 → 59 + 1 = 60
           # x = 60 → 60 + 1 = 61
           # x = 45 → 45 + 1 = 46



```

**Ordenación de datos:**

```python
df=df.sort_values(by="Edad", ascending=False)
df
```

**Selección de columnas:**`

```python
columnas_seleccionadas=df[["Nombre","Genero"]]
columnas_seleccionadas   #Dataframe nuevo a partir de otro
```

**Agregación y agrupación de datos:**

```python
df["Ciudad"].groupby([df.Ciudad]).agg("count")
```

```python
df["Ciudad"].groupby([df.Ciudad]).agg("count").plot(kind="pie",fig=(5,4))
```

```python
df["Ciudad"].groupby([df.Ciudad]).agg("count").plot(kind="bar",fig=(5,4))
```

**Limpieza de datos nulos:**

```python
df=df.dropna()
```

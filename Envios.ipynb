```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar el archivo CSV
envios=pd.read_csv("/content/seguimiento_envios.csv", sep=',')

# Porcentaje de envíos retrasados
total_envios = len(envios)
envios_retrasados = len(envios[envios["Estado_Envio"] == "Retrasado"])
porcentaje_retrasados = (envios_retrasados / total_envios) * 100
print("Porcentaje de envíos retrasados:", porcentaje_retrasados, "%")

# Ciudades con más retrasos
ciudades_retrasos = envios[envios["Estado_Envio"] == "Retrasado"]["Ciudad_Destino"].value_counts()
print("Ciudades con más retrasos:\n", ciudades_retrasos)

# Gráfico de barras: Ciudades con retrasos
plt.figure(figsize=(8, 5))
ciudades_retrasos.plot(kind="bar", color="salmon")
plt.title("Ciudades con Más Retrasos en Entregas")
plt.xlabel("Ciudad Destino")
plt.ylabel("Cantidad de Retrasos")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

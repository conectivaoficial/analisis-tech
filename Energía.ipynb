```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Datos ficticios
data = {
    'año': [2018, 2018, 2018, 2019, 2019, 2019, 2020, 2020, 2020],
    'fuente_energía': ['solar', 'eólica', 'hidroeléctrica', 'solar', 'eólica', 'hidroeléctrica', 'solar', 'eólica', 'hidroeléctrica'],
    'generación_mwh': [500, 300, 1000, 600, 350, 1100, 700, 400, 1200],
    'consumo_mwh': [1500, 1500, 1500, 1600, 1600, 1600, 1700, 1700, 1700]
}

df = pd.DataFrame(data)
```

```
# Visualización de Resultados
tabla_pivot = df.pivot(index='año', columns='fuente_energía', values='generación_mwh')
sns.heatmap(tabla_pivot, annot=True, fmt='.0f', cmap='YlGnBu')
plt.title('Generación de Energía por Fuente y Año')
plt.show()
```

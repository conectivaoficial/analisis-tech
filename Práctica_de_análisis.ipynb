Carga y Almacenamiento de Datos:

Pandas ofrece una variedad de métodos para cargar y almacenar datos desde y hacia diferentes fuentes, como archivos CSV, Excel, bases de datos SQL y más. Esto facilita la importación y exportación de datos para su posterior análisis.

Ejemplo de carga de datos desde un archivo CSV:

```python
import pandas as pd
```

```python
cajamenor=pd.read_csv("/content/cajamenor.csv", sep=",")
cajamenor
```

```python
cajamenor.info()   #informacion general de mi dataframe
```

```python
cajamenor.columns  #nombre de mis columnas en mi dataframe
```

```python
cajamenor.columns=['Concepto', 'Fecha', 'Entrada', 'Salida', 'Categoría', 'Proveedor','Factura', 'Saldo']  #editando los nombres de mis columnas
cajamenor.columns
```

```python
cajamenor["prueba"]=34   #Creando una columna nueva de prueba
cajamenor
```

```python
del(cajamenor["prueba"]) #Eliminar columna
cajamenor
```

```python
cajamenor["Fecha"]=pd.to_datetime(cajamenor["Fecha"], format="%d/%m/%Y")   #16/05/2021  Conversion de campos tipo fecha
cajamenor
```

```python
cajamenor.info()
```

```python
cajamenor["Fecha"].groupby([cajamenor.Fecha.dt.day]).agg("count").sort_values(ascending=False)
```

```python
cajamenor["Categoría"].groupby([cajamenor.Categoría]).agg("count").sort_values(ascending=False)
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar el archivo CSV
ventas=pd.read_csv("/content/ventas_productos.csv", sep=',')

ventas["Fecha_Venta"] = pd.to_datetime(ventas["Fecha_Venta"])
# Producto más vendido
producto_mas_vendido = ventas.groupby("Producto")["Cantidad"].sum().idxmax()
print("Producto más vendido:", producto_mas_vendido)

# Días de la semana con más ventas
ventas["Día_Semana"] = ventas["Fecha_Venta"].dt.day_name()
ventas_por_dia = ventas.groupby("Día_Semana")["Cantidad"].sum()
print("Días de la semana con más ventas:\n", ventas_por_dia)

# Gráfico de barras: Ventas por día de la semana
plt.figure(figsize=(8, 5))
ventas_por_dia = ventas_por_dia.reindex(["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])
ventas_por_dia.plot(kind="bar", color="skyblue")
plt.title("Ventas por Día de la Semana")
plt.xlabel("Día de la Semana")
plt.ylabel("Cantidad Vendida")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```
